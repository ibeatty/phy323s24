# Solutions for Problem Set S2

## Problem 4: Taylor 2.43

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
m = 0.6
gamma = 0.25
D = 0.24
c = gamma * D**2
c_m = c/m
g = 9.81

x, y = 0.0, 2.0
x0, y0 = x, y
v_i, q_i = 15.0, np.deg2rad(45.0)
vx, vy = v_i * np.cos(q_i), v0 * np.sin(q_i)
vx0, vy0 = vx, vy
dt = 0.01

x_vals,  y_vals = [x,], [y,]
x0_vals,  y0_vals = [x,], [y,]
while y >= 0.0 or y0 >= 0.0:
    v = np.sqrt(vx**2 + vy**2)
    ax = -c_m * v * vx
    ay = -c_m * v * vy - g
    ax0 = 0.0
    ay0 = -g
    vx += ax * dt
    vy += ay * dt
    vx0 += ax0 * dt
    vy0 += ay0 * dt
    x += vx * dt
    y += vy * dt
    x0 += vx0 * dt
    y0 += vy0 * dt
    x_vals.append(x)
    y_vals.append(y)
    x0_vals.append(x0)
    y0_vals.append(y0)

fig, ax = plt.subplots()
ax.plot(x_vals, y_vals, label="with drag")
ax.plot(x0_vals, y0_vals, label="in vacuum")
ax.set_xlabel('x [m]')
ax.set_ylabel('y [m]')
ax.set_xlim(left = 0)
ax.set_ylim(bottom = 0)
ax.grid()
ax.legend()
ax.set_title("Basketball's trajectory with and without drag")
plt.show()
```
